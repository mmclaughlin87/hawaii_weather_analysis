```python
import pandas as pd
import os
```

```python
measurement_file = os.path.join("Resources", "hawaii_measurements.csv")
station_file = os.path.join("Resources", "hawaii_stations.csv")

measurement_data = pd.read_csv(measurement_file)
station_data = pd.read_csv(station_file)
```

```python
print(f"Total rows: {measurement_data.count()}")
print(f"Null values: {measurement_data.isnull().sum()}")
measurement_data.head(10)
```

```python
print(f"Total rows: {station_data.count()}")
print(f"Null values: {station_data.isnull().sum()}")
station_data.head(10)
```

```python
clean_measurement_data = measurement_data.dropna(how="any") 
clean_station_data = station_data

measurement_save_path = os.path.join("Resources", "clean_hawaii_measurements.csv")
station_save_path = os.path.join("Resources", "clean_hawaii_stations.csv")

clean_measurement_data.to_csv(measurement_save_path, index=False, header=True)
clean_station_data.to_csv(station_save_path, index=False, header=True)
```
